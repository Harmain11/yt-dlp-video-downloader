```python
import yt_dlp

url = 'https://www.empwrfitnessstudio.com/physical-therapy'  # The URL of the page

ydl_opts = {
    'outtmpl': '%(title)s.%(ext)s',  # Save the video with its title
    'format': 'best',  # Choose the best available quality
}

with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    ydl.download([url])
```
